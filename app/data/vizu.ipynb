```python
import seaborn as sns
import pandas as pd
%matplotlib inline
```

```python
infile = 'train_users_2.csv'
df = pd.read_csv(infile)
df_bis = df[['age', 'gender', 'country_destination']].dropna()
df_plot = df_bis[(df['age'] >= df['age'].quantile(0.1)) & (df['age'] <= df['age'].quantile(0.9))]
```

```python
df_plot
```

```python
sns.pairplot(df_plot, hue='country_destination')
```
